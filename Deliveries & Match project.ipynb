```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
cric_data = pd.read_csv(r'C:\Users\patel d d\Desktop\project python files\deliveries cricket Data.csv')
cric_data
```

```python
cric_data.shape
```

```python
cric_data.info()
```

```python
cric_data.isnull().sum()
```

As we can see above, dataset has many missing values so it needs to be handled but in our case its match innegings data so this missing values make sense we will not apply data imputation method on it.

```python
run_total_match = cric_data.groupby(['match_id','inning','batting_team'])['total_runs'].sum()
run_total_match
```

```python
run_total_match = cric_data.groupby(['match_id','inning','batting_team'])['total_runs'].sum()
winners = run_total_match.groupby('match_id').max()
winners
```

* # winner teams

```python
winner_team = run_total_match.groupby('match_id').idxmax()
winning_team = winner_team.apply(lambda x : x[2])
winning_team
```

```python
count_winning_team = winning_team.value_counts()
count_winning_team
fig, ax = plt.subplots(figsize = (12,6))
count_winning_team.plot(kind = 'bar',color = 'skyblue')
plt.xlabel('IPL TEAMS')
plt.ylabel('MATCHES')

plt.title('WINNING TEAM')
plt.grid(axis='y', linestyle='--')
plt.show()
```

* # Mumbai indians is max time winner


* # ORANGE CAPE HOLDER 

```python
max_run = cric_data.groupby('batsman')['total_runs'].sum()
best_batsman = max_run.sort_values(ascending=False)
best_batsman = best_batsman.head(10)
best_batsman

plt.figure(figsize=(10,6))
best_batsman.plot(kind='line',color= 'orange')
plt.xlabel('Batsmans')
plt.ylabel('Total runs')
plt.xticks(rotation=90)
plt.title('Best Batsman by Runs')
plt.grid(axis='y')
plt.show()
```

* # Most successful bowler in term of wicket taker
* # PURPLE CAPE HOLDER


```python
wicket_taker = cric_data.groupby('bowler')['player_dismissed'].count()
wicket_taker = wicket_taker.sort_values(ascending=False)
wicket_taker = wicket_taker.head(10)
wicket_taker
plt.figure(figsize=(12,6))
wicket_taker.plot(kind='line',color= 'purple')
plt.xlabel('BOWLER')
plt.ylabel('TOTAL WICKETS')
plt.title('BEST BOWLER IN TERMS OF WICKET TAKEN')
plt.grid(axis='y')
plt.show()
```

* # Top batsmen with highest strike rate

```python
# strike rate : run / balls 
batsman_stike_rate = cric_data.groupby('batsman')['total_runs'].sum() / cric_data.groupby('batsman')['ball'].sum()
batsman_stike_rate = batsman_stike_rate.sort_values(ascending=False).head(10)
batsman_stike_rate

plt.figure(figsize=(12,6))
batsman_stike_rate.plot(kind='bar',color= 'yellow')
plt.xlabel('BATSMANS')
plt.ylabel('STRIKE RATES')
plt.title('TOP BATSMAN WITH HIGHEST STRIKE RATE')
plt.grid(axis='y')
plt.show()
```

* # EXTRA RUNS conceded by each team 

```python
extra_conceded = cric_data.groupby('bowling_team')['extra_runs'].sum().sort_values(ascending=True)
extra_conceded
plt.figure(figsize=(12,6))
extra_conceded.plot(kind='bar',color= 'green')
plt.xlabel('IPL TEAMS')
plt.ylabel('EXTRA RUNS')
plt.title('EXTRA RUNS CONCEDED BY EACH TEAMS')
plt.grid(axis='y')
plt.show()
```

* # Highest partnership by runs

```python
partnership = cric_data.groupby(['match_id','batsman','non_striker'])['total_runs'].sum().sort_values(ascending=False).head(10)
partnership
```

* # Dismissal type by batsman 

```python
Dismissal_type_player = cric_data.groupby(['batsman','dismissal_kind'])['player_dismissed'].count().reset_index()
Dismissal_type_player.pivot(index='batsman',columns = 'dismissal_kind',values = 'player_dismissed')
```

* # Analyze powerplay score

```python
powerplay_score = cric_data[cric_data['over']<= 6].groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
powerplay_score
plt.figure(figsize=(12,6))
powerplay_score.plot(kind='bar',color= 'PINK')
plt.xlabel('BATTING TEAMS')
plt.ylabel('POWERPLAY TOTAL RUNS')
plt.title('POWERPLAY SCORE')
plt.grid(axis='y')
plt.show()
```

* # Top bowlers in term of economy rate

```python
bowler_economy = cric_data.groupby('bowler')['total_runs'].sum() / cric_data.groupby('bowler')['over'].sum()
bowler_economy = bowler_economy.sort_values(ascending=False).head(10)
bowler_economy
plt.figure(figsize=(12,6))
bowler_economy.plot(kind='bar',color= 'orange')
plt.xlabel('BOWLER')
plt.ylabel('ECONOMY RATE')
plt.title('TOP BOWLERS IN TERM OF ECONOMY RATE')
plt.grid(axis='y')
plt.show()
```

* # Top 10 fielders in tearms of catches taken

```python
fielder_catches = cric_data[cric_data['dismissal_kind'] == 'caught' ].groupby(['fielder','dismissal_kind'])['fielder'].count()
fielder_catches = fielder_catches.sort_values(ascending=False).head(10)
fielder_catches
plt.figure(figsize=(12,6))
fielder_catches.plot(kind='bar',color= 'YELLOW')
plt.xlabel('FIELDER')
plt.ylabel('CATCH SCORES')
plt.title('FIELDERS IN TEARMS OF CATCH TAKEN')
plt.grid(axis='y')
plt.show()
```

* # TEAMS TOSS DECISIONS

```python
teams_toss_decision = match_data.pivot_table(columns='toss_decision',index='toss_winner',aggfunc='size', fill_value=0)
teams_toss_decision
teams_toss_decision.plot(kind='bar')
plt.figure(figsize=(12, 6))
plt.title('TEAMS TOSS DECISIONS')
plt.xlabel('Teams')
plt.xticks(rotation=90)
plt.show()
```

* # win by max runs margin

```python
wins_by_run = match_data.groupby('winner')['win_by_runs'].max().sort_values(ascending=False)
wins_by_run
```

* # win by wickets


```python
win_wickets = match_data.groupby('winner')['win_by_wickets'].max().sort_values(ascending=False)
win_wickets
```
